**Imbalanced Dataset :-**
A dataset with an unequal class distribution

```python
import numpy as np
import pandas as pd
```

```python
#loading the dataset to pandas dataset
credit_card_data= pd.read_csv('/content/credit_data.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
#determine the distribution of two classes
credit_card_data['Class'].value_counts()
```

This is highly imbalanced dataset

0 ->>>> Legit transaction
1 ->>>> fraudelent transactions

```python
#separating the legit and farudelent transactions
legit = credit_card_data[credit_card_data.Class == 0 ]
fraud = credit_card_data[credit_card_data.Class == 1 ]
```

```python
print(legit.shape)
print(fraud.shape)
```

under-sampling >> technique to handle imbalanced data

Building the sample dataset containing similar distribution of legit and fraudulent transactions

Number of fraudelent transactions -->> 81

```python
legit_sample = legit.sample(n=81)
```

```python
print(legit_sample.shape)
```

concatenate the two dataframes

```python
new_dataset= pd.concat([legit_sample , fraud] , axis = 0)
```
